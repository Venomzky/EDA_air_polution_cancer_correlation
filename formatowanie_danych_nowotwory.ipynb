```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("./data/Zachorowania1999-2018_wojewodztwa.csv", sep=';')
data = data.dropna()
data.rename(columns={'Rok (rok)':'Rok','Wojewodztwo (wojewodztwo)':'Województwo','Plec (plec)':'plec','ICD10 (icd10)':'icd10','Liczba (liczba)':'Liczba','Wiek (wiek)':'wiek'}, inplace=True)
data['Rok'] = data['Rok'].astype(int)
data['Województwo'] = data['Województwo'].astype(int)
data['Liczba'] = data['Liczba'].astype(int)
mapowanie = {
    2: 'Dolnośląskie',
    4: 'Kujawsko-Pomorskie',
    8: 'Lubuskie',
    10: 'Łódzkie',
    6: 'Lubelskie',
    12: 'Małopolskie',
    14: 'Mazowieckie',
    16: 'Opolskie',
    20: 'Podlaskie',
    18: 'Podkarpackie',
    22: 'Pomorskie',
    26: 'Świętokrzyskie',
    24: 'Śląskie',
    28: 'Warmińsko-Mazurskie',
    30: 'Wielkopolskie',
    32: 'Zachodniopomorskie'
}
data['Województwo'] = data['Województwo'].map(mapowanie)
data = data[data['wiek'] != '0-4']
data
```

```python
#nowotwory z powodu Benzenu (c6h6)
c6h6_cancer = data[(data['icd10']<='C96' ) & (data['icd10']>='C81') | (data['icd10']=='C49')]
c6h6_cancer = c6h6_cancer.groupby(['Rok','Województwo'])['Liczba'].sum().reset_index()
c6h6_cancer.to_csv('./data/c6h6_nowotwory.csv', index=False)
c6h6_cancer[:5]
```

```python
#nowotwory z powodu wwa (Wielopierścieniowe węglowodory aromatyczne)
wwa_cancer = data[(data['icd10']>='C30' ) & (data['icd10']<='C39') | (data['icd10']=='C43') | (data['icd10']=='C44') | (data['icd10']<='C22' ) & (data['icd10']>='C18')]
wwa_cancer = wwa_cancer.groupby(['Rok','Województwo'])['Liczba'].sum().reset_index()
wwa_cancer.to_csv('./data/wwa_nowotwory.csv', index=False)
wwa_cancer[:5]
```

```python
#nowotwory z powodu pm2,5
pm2_5_cancer = data[(data['icd10']>='C30' ) & (data['icd10']<='C39')]
pm2_5_cancer = pm2_5_cancer.groupby(['Rok','Województwo'])['Liczba'].sum().reset_index()
pm2_5_cancer.to_csv('./data/pm2_5_nowotwory.csv', index=False)
pm2_5_cancer[:5]
```

```python
lungs_cancer = data[data['icd10'] =='C34']
lungs_cancer = lungs_cancer.groupby(['Rok','Województwo'])['Liczba'].sum().reset_index()
lungs_cancer.to_csv('./data/pluca_nowotwory.csv', index=False)
lungs_cancer[:5]
```

```python
grouped_data = data.groupby(['Rok','Województwo'])['Liczba'].sum().reset_index()
grouped_data.to_csv('./data/nowotwory.csv', index=False)
```
