### Zacznijmy od pm2_5 wyświetlmy zachorowania z wszystkich dotępnych lat dla województw oraz stężenie pm2,5 w powietrzu.

```python
data_merged = pd.merge(pm2_5, nowotwory_pm2_5, on=['Rok', 'Województwo'])
data_merged_a = data_merged[data_merged['Województwo'] > 'Podkarpackie']
data_merged_b = data_merged[data_merged['Województwo'] <= 'Podkarpackie']

plt.figure(figsize=(10, 8))
for wojewodztwo, group in data_merged_a.groupby('Województwo'):
    plt.scatter(group['Średnia'], group['Liczba'], label=wojewodztwo, alpha=0.5)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))


    
plt.xlabel('stęzenie pm2.5 w powietrzu(µg/m3)')
plt.ylabel('Liczba nowotoworów')
plt.title('pm2,5 a zachorowania dla województw')
plt.show()

plt.figure(figsize=(10, 8))
for wojewodztwo, group in data_merged_b.groupby('Województwo'):
    plt.scatter(group['Średnia'], group['Liczba'], label=wojewodztwo, alpha=0.5)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    
plt.xlabel('stęzenie pm2.5 w powietrzu(µg/m3)')
plt.ylabel('Liczba nowotoworów')
plt.title('pm2,5 a zachorowania dla województw')
plt.show()
```

Nie da się wskazać ogólnej zależności. Dla jednych województwo wygląda jakby dało dostrzec się korelacje między stęzeniem pm2.5 w powietrzu a zachorowaniami na nowotwory w innych jest to nie możliwe.

### Sprawdzenie korelacji między stężeniem pm2,5 w powietrzu a nowotworami przy czym wyniki nowotworów są przesuwane w czasie. Próba znalezienia wplywu pm2,5 na nowotwory dopiero w nastepnych latach.

```python
heat_data_pm2_5 = []
wojewodztwa = []
for i in range(11):
    nowotwory_pm2_5_rok = nowotwory_pm2_5.copy()
    nowotwory_pm2_5_rok['Rok'] = nowotwory_pm2_5_rok['Rok']+i
    data_merged = pd.merge(pm2_5, nowotwory_pm2_5_rok, on=['Rok', 'Województwo'])
    row = []
    for wojewodztwo, group in data_merged.groupby('Województwo'):
        corr = group['Średnia'].corr(group['Liczba'])
        row.append(corr)
    heat_data_pm2_5.append(row)
heat_data_pm2_5 = np.array(heat_data_pm2_5)
heat_data_pm2_5 = np.transpose(heat_data_pm2_5)

for wojewodztwo, group in data_merged.groupby('Województwo'):
    wojewodztwa.append(wojewodztwo)
```

```python
sns.set(style="white")
plt.figure(figsize=(10, 8))
sns.heatmap(heat_data_pm2_5, annot=True, cmap="coolwarm", fmt=".2f",
            xticklabels=list(range(0, 11)),
            yticklabels=wojewodztwa)
plt.title('Kolrelacja pomiarów a nowotworami 0-10 lat w przyszłości')
plt.show()
```

Nie udało się zauważyć charakterystycznego wpływu pm2_5 na ilość zachorowań na nowotwory w przyszłości. Sytuacja ma się identycznie dla innych pierwiastków, jednak ograniczę się do jednego wykresu.

Ciekawie wygląda województwo Małopolskie, sprawdźmy czy uda się coś zaobserwować. Dla wojwództwa Małopolskiego sprawdzimy dane dla czterech najgroźniejszych pierwiastków.

```python
data_merged = pd.merge(pm2_5, wwa, on=['Rok', 'Województwo'])
data_merged.rename(columns={'Średnia_x':'pm2_5', 'Średnia_y':'wwa'}, inplace=True)
data_merged = pd.merge(data_merged, NOx, on=['Rok', 'Województwo'])
data_merged.rename(columns={'Średnia':'NOx', 'Średnia_y':'wwa'}, inplace=True)
data_merged = pd.merge(data_merged, C6H6, on=['Rok', 'Województwo'])
data_merged.rename(columns={'Średnia':'C6H6', 'Średnia_y':'wwa'}, inplace=True)
ret = data_merged.copy()
data_merged = data_merged[data_merged['Województwo'] == 'Małopolskie']
data_merged.drop('Województwo', axis=1, inplace=True)
```

```python
data_merged.plot(x='Rok', kind='bar', stacked = True)
plt.ylabel('pm2.5 w powietrzu(µg/m3)')
plt.title('Stężenie pierwiastków Małopolskie')
plt.show()
```

Możemy zaobserwować, że stężenie niebezpiecznych pierwiastków w powietrzu maleje na przestrzeni lat.

```python
nowotwory_malopolskie_wszystkie = nowotwory[nowotwory['Rok'] >= 2009]
nowotwory_malopolskie_wszystkie = nowotwory_malopolskie_wszystkie[nowotwory_malopolskie_wszystkie['Województwo']=='Małopolskie']
nowotwory_malopolskie_C6H6 = nowotwory_C6H6[nowotwory_C6H6['Rok'] >= 2009]
nowotwory_malopolskie_C6H6 = nowotwory_malopolskie_C6H6[nowotwory_malopolskie_C6H6['Województwo']=='Małopolskie']
nowotwory_malopolskie_p2_5 = nowotwory_pm2_5[nowotwory_pm2_5['Rok'] >= 2009]
nowotwory_malopolskie_p2_5 = nowotwory_malopolskie_p2_5[nowotwory_malopolskie_p2_5['Województwo']=='Małopolskie']
nowotwory_malopolskie_p2_5
lata = list(range(2009,2019))
plt.figure(figsize=(10,10))
plt.xlim(min(lata), max(lata))
line1, = plt.plot(lata, nowotwory_malopolskie_p2_5['Liczba'])
line2, =plt.plot(lata, nowotwory_malopolskie_C6H6['Liczba'])
line3, =plt.plot(lata, nowotwory_malopolskie_wszystkie['Liczba'])
plt.legend(handles=[line1, line2, line3], labels=['nowotwory charakterysztyczne dla p2,5', 'nowotwory charakterysztyczne dla C6H6', 'wszystkie typy nowotworów'])
plt.title('Liczba nowotworów województwo Małopolskie')
plt.xlabel('Rok')
plt.ylabel('Liczba nowotworów')
plt.show()
```

Stężenie pierwiastków maleje z roku na rok, liczba nowotworów jednak nie naśladuje tego trendu.

### Zbadajmy jeszcze zróżnicowanie pomiarów w skali całego kraju, oraz liczbę nowotowrów tylko płuc.

```python
data_merged = pd.merge(pm2_5, wwa, on=['Rok', 'Województwo'])
data_merged.rename(columns={'Średnia_x':'pm2_5', 'Średnia_y':'wwa'}, inplace=True)
data_merged = pd.merge(data_merged, NOx, on=['Rok', 'Województwo'])
data_merged.rename(columns={'Średnia':'NOx', 'Średnia_y':'wwa'}, inplace=True)
data_merged = pd.merge(data_merged, C6H6, on=['Rok', 'Województwo'])
data_merged.rename(columns={'Średnia':'C6H6', 'Średnia_y':'wwa'}, inplace=True)
data_merged['suma'] = data_merged['pm2_5'] + data_merged['wwa'] + data_merged['NOx'] + data_merged['C6H6']
data_merged.drop(['Województwo','pm2_5','wwa','NOx','C6H6'], axis=1, inplace=True)
data_merged = data_merged[data_merged['Rok']>2008]
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='suma', y='Rok', data=data_merged, orient='h')
plt.xlabel('stężenie pierwiastków (µg/m3)')
plt.ylabel('Rok')
plt.title('Suma stężeń pierwiastków chemicznych')

# Pokazanie wykresu
plt.show()
```

Widzimy, że średnie stężenie związków chemicznych z roku na rok maleje, zmniejsza się również wariacja pomiarów.

```python
data_merged
srednie_zanieczyszczenia = data_merged[data_merged['Rok']>2008]
srednie_zanieczyszczenia = srednie_zanieczyszczenia.groupby('Rok')['suma'].mean()
srednie_zanieczyszczenia
nowotwory_tylko_pluca = nowotwory_pluc[nowotwory_pluc['Rok'] > 2008]
nowotwory_tylko_pluca = nowotwory_tylko_pluca.groupby('Rok')['Liczba'].sum()
nowotwory_tylko_pluca.plot(kind='line')
plt.title('Liczba nowtworów płuc Polska')
plt.xlabel('Rok')
plt.ylabel('Liczba nowotworów')
plt.show()
srednie_zanieczyszczenia.plot(kind='line')
plt.title('Średnie zanieczyszczenie Polska')
plt.xlabel('Rok')
plt.ylabel('stężenie pierwiastków (µg/m3)')
plt.show()
```

Na obrazie całego kraju możemy dostrzec, że mimo zmniejszającego się stężenia związków chemicznych liczba nowotworów nie ulega zmniejszeniu się.
