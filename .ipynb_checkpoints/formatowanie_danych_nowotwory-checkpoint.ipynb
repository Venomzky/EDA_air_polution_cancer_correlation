```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("./data/Zachorowania1999-2018_wojewodztwa.csv", sep=';')
data = data.dropna()
data.rename(columns={'Rok (rok)':'rok','Wojewodztwo (wojewodztwo)':'wojewodztwo','Plec (plec)':'plec','ICD10 (icd10)':'icd10','Liczba (liczba)':'liczba','Wiek (wiek)':'wiek'}, inplace=True)
data['rok'] = data['rok'].astype(int)
data['wojewodztwo'] = data['wojewodztwo'].astype(int)
data['liczba'] = data['liczba'].astype(int)
mapowanie = {
    2: 'Dolnośląskie',
    4: 'Kujawsko-Pomorskie',
    8: 'Lubuskie',
    10: 'Łódzkie',
    6: 'Lubelskie',
    12: 'Małopolskie',
    14: 'Mazowieckie',
    16: 'Opolskie',
    20: 'Podlaskie',
    18: 'Podkarpackie',
    22: 'Pomorskie',
    26: 'Świętokrzyskie',
    24: 'Śląskie',
    28: 'Warmińsko-Mazurskie',
    30: 'Wielkopolskie',
    32: 'Zachodniopomorskie'
}
data['wojewodztwo'] = data['wojewodztwo'].map(mapowanie)
data
```

```python
grouped_data = data.groupby(['rok','wojewodztwo'])['liczba'].sum().reset_index()
```

```python
grouped_data.to_csv('./data/nowotwory.csv', index=False)
```
